# Data preprocessing 

```python
# Change the Categorical column to 0, 1
categorical_column =data[['mainroad','guestroom','basement','hotwaterheating','airconditioning','furnishingstatus']]

le =LabelEncoder()
label_encoders  = {col: le for col in categorical_column}

for col in categorical_column:
    data[col] = label_encoders[col].fit_transform(data[col])
    
data.head()
```

```python
# Normalize numerical columns (Simplify large numbers)
numerical_column = data[['price', 'area' ,'parking','bathrooms','bedrooms']]
scaler = StandardScaler()
data[numerical_column.columns] = scaler.fit_transform(numerical_column)
data.head()
```

# Create model

```python
# Split Data for features (X) and target (y)
X = data.drop(['price'], axis=1)
y = data['price']
```

```python
# Cost function
def compute_cost(X, y, w, b):
    m = X.shape[0]
    f_wb = np.dot(X, w) + b 
    errors = f_wb - y
    j_wb = (1 / (2 * m)) * np.sum(errors ** 2) 
    return j_wb
```

```python
# Gradient Descent
def gradient_descent(X, y, w, b, alpha, num_iters):
    m = X.shape[0]
    cost_history = []

    for _ in range(num_iters):
        f_wb = np.dot(X, w) + b  
        cost = f_wb - y
        dw = (1 / m) * (X.T @ cost)
        db = (1 / m) * np.sum(cost)
        w -= alpha * dw
        b -= alpha * db
        cost_history.append(compute_cost(X, y, w, b))

    return w, b, cost_history
```

```python
# Create model 
def model(X, y, alpha=0.01, num_iters=1100):
    y = y.to_numpy().reshape(-1, 1)  
    w = np.zeros((X.shape[1], 1))
    b = 0

    w, b, cost_history = gradient_descent(X, y, w, b, alpha, num_iters)

    def predict(X_new):
        f_wb = (X_new @ w + b).flatten()
        return f_wb

    return w.flatten(), b, cost_history, predict
```

```python
# Traing the model
X_with_intercept = np.hstack([np.ones((X.shape[0], 1)), X])  
w, b, cost_history, predict = model(X_with_intercept, y, alpha=0.01, num_iters=1100)
```

```python
# Show Cost Function
plt.figure(figsize=(10, 4))
plt.plot(cost_history)
plt.title("Cost Function")
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.show()
```

```python
X_new = np.array([[-1.171756, 1.403419, 3.41381, 0, 0, 0, 0, 0, -0.805741, 1]])
X_new = X_new.astype(float) 

X_new_with_intercept = np.hstack([np.ones((X_new.shape[0], 1)), X_new]) 

predictions = predict(X_new_with_intercept)
print(predictions)
```
